###### Ether, lumière, espace

Abandonner le concept d'éther ne fut certainement pas la meilleure idée scientifique. Que contient l'espace ? 

L'éther est l'impédance du vide Z ou R (impédance et résistance), la métaphore de l'espace est l'inductance ($L$) et la capacitance ($C$) dont le rapport donne, (a) : l'impédance (l'éther en terme matéphysique),(b) : La fréquence angulaire, la pulsation fondamentale de la lumière.

La terminalogie des variables peut s'échanger sans en changer la valeur physique.

Inductance : $L=\mu$ et capacitance $C=\epsilon$
, l'impédance $Z=\sqrt\frac{L}{C}=Z=\sqrt\frac{\mu}{\epsilon}$, en modifiant l'impédance (l'éther) on modifie également la fréquence angulaire et de là, la polarisation de l'onde $\lambda$.

Ainsi la lumière est le comma ou coma ($ç$ pour ne pas confondre avec $c$, puisque c'est une rupture de la circonférence, la circonférence = $ç=\pi 10\times ^8 =314159265$ et la célérité  $c=299792458$
