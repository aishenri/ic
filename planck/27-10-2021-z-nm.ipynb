```python
c=299792458;mu=n(pi)*4E-7;ep=1/(mu*c*c)
c=1/(sqrt(mu*ep));round(c)
```

```python
var("omega")
R=0.00376;L=0.001;C=0.885;w=round(1/(sqrt(L*C)),2);
p=plot(abs(1/(1+i*omega*R*C-omega*omega*L*C)),
       [omega,0,2*w],color="darkgreen")
f=w;r=R;c=C;l=L
lbl = text("$w_r = %.0f rad/s$" % f,(4,10),
           rgbcolor="darkgreen",
           horizontal_alignment="left",
           fontsize=10,
           bounding_box={'boxstyle':'round',\
                         'fc':'w'})
show(p+lbl,figsize=3,frame=True,axes=False,
     gridlines="minor",fontsize=6,
     gridlinesstyle=dict(color="lightsalmon",
                         linestyle="-",alpha=0.4,
                         linewidth="0.5"))
```
