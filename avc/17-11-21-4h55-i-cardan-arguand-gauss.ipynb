Gauss : $\sqrt-1$ pose un problème ? (ou pas ?)

On abouti à une contradiction :

$\sqrt a \sqrt b=\sqrt ab$

$\sqrt -1 \sqrt -1=(\sqrt -1)^2=-1$

$\sqrt -1 -1 =\sqrt(-1)^2=\sqrt1=1$

L'opposé d'un nombre a est l'unique nombre b tel que a+b=0; il se note - a

Il n'y a pas contradiction, les parenthèses donne un  résultat opposé. Avec un logiciel on ne rencontre pas celle-ci : Opposé de 1 = -1

Pourtant il reste un problème obscure : $\sqrt -1=i$ a-t-il une autre valeur réelle décimale, en d'autre terme : Quelle est sa position fondamentale quand $|z|$ ?

L'inverse d'un nombre a non nul est l'unique nombre b tel que a×b=1 ; il se note $\frac{1}{a}$ 

$z=a+b.i$  et $|z|=\sqrt a²+b²$ (Où est $i$ ?)
