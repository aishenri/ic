En musique, un comma est un intervalle très petit, généralement entre le dixième et le cinquième du ton. Il est défini comme étant l’intervalle entre deux séquences d’intervalles purs.

Cet intervalle correspond approximativement à l'écart de fréquence entre un la à 440 Hz et un la à 446 Hz, soit 6 battements par seconde. 

Il a pour rapport acoustique 312⁄219 = 531441⁄524288. Il vaut environ 23,46 cents.

La fréquence d'une note est multipliée par 1,5 en la montant d'une quinte pure et par 2 en la montant d'une octave. Au bout de 12 quintes on tombe sur la note enharmonique de celle obtenue au bout de 7 octaves. La formule ci-dessus s'explique donc par l'égalité suivante :
